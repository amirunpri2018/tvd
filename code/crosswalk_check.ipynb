```python
# calculation to use crosswalk as a two horizontal lines
top_of_xwalk = xwalk[1]- xwalk[3]/2
bot_of_xwalk = xwalk[1] + xwalk[3]/2
```

### Functions

```python
# Look for crossingwalk violation using the below criteria:
# Mininmum of 4 consecutive frames with a don't walk light detected
# A car within the crosswalk (past a threshold)
# Don't walk lights are object #1
# Each file is a frame equal to 1 second, so check neighboring files
# for criteria
# (0,0) is the TOP-LEFT corner
def xwalk_on_red(frame):
    detections = pd.read_csv(frame, sep=' ', header=None)
    detections.set_axis(['object', 'x', 'y', 'width', 'height'], axis=1, inplace=True)

    # Here's the math to test against basic rectangle crosswalk
    detections['y_bottom'] = (detections['y']+(detections['height']/2))
    detections['y_dist_to_xwalk_top'] = top_of_xwalk-detections['y_bottom']
    detections['y_dist_to_xwalk_bot'] = bot_of_xwalk-detections['y']
      
    # is the bottom of the car below the top of the xwalk?
    below_xwalk = (detections[detections['y_dist_to_xwalk_top']<-0.01])
    
    # is the center of the car below the bottom of xwalk?
    near_xwalk = (below_xwalk[below_xwalk['y_dist_to_xwalk_bot']<-.01])
    
    # is the car vertically aligned?
    vert_below = near_xwalk[near_xwalk['width']<near_xwalk['height']]
    
    # check if a car is in the xwalk and signal is red
    if (len(vert_below)>0) and\
        (len(detections[detections['object']==1])>0):
            return(1) #signifies suspicion in this frame
    else:
        return(0) #signifies no suspicion in this frame
```

### Main body

```python
# Check everything in the directory
results = list(enumerate([xwalk_on_red(item) for item in os.listdir()]))
```

```python
# test for consecutive runs of n length of frames of suspicion 
frames_of_interest = []
for i in range(0,len(results)-n):
    maybes = 0
    for j in range(n):
        if results[i+j][1]==1:
            maybes += 1        
    if maybes == n:
        frames_of_interest.append(i)
```
